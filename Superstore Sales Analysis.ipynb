# Exploratory Data Analysis

## Customer Analysis

### Customer segmentation

```python
df.head(3)
```

```python
# Types of customer

types_of_customer = df['Segment'].unique()
print(types_of_customer)
```

```python
# Number of customer in each segment

number_of_customers = df['Segment'].value_counts().reset_index()
number_of_customers = number_of_customers.rename(columns={'count':'Total Customers', 'Segment':'Customer Type'})

print(number_of_customers)
```

## Customers and Sales

```python
# Bar Graph

plt.bar(sales_per_category['Customer Type'], sales_per_category['Total Sales'])

# Lbael
plt.title('Sales per Customers Category')
plt.xlabel('Customer Type')
plt.ylabel('Total Sales')

plt.show()
```

```python

plt.pie(sales_per_category['Total Sales'], labels=sales_per_category['Customer Type'], autopct='%1.1f%%')

# Set the title of the pie chart
plt.title('Sales per Customer Category')
plt.show()
```

## Customer Loyalty
* Examine the repeat purchase behaviour of customers

```python
df.head(3)
```

```python
# Group data according to customer ID,Customer Name, Segment and calculate freq. of their orders
customer_order_freq = df.groupby(['Customer_ID', 'Customer_Name', 'Segment'])['Order_ID'].count().reset_index()

# Rename the order ID Column
customer_order_freq.rename(columns={'Order_ID': 'Total Orders'}, inplace=True)

# Identify repeat customers
repeat_customers = customer_order_freq[customer_order_freq['Total Orders']>1]

# sort repeat customer in descending order
sorted_repeat_customers = repeat_customers.sort_values(by='Total Orders', ascending=False)

print(sorted_repeat_customers.head(10).reset_index(drop=True))
```

```python
# Group data based on : Custromer ID, Customer Name and Sales

customer_sales = df.groupby(['Customer_ID', 'Customer_Name'])['Sales'].sum().reset_index()

# Rename the order ID Column
customer_sales.rename(columns={'Customer_ID': 'Customer ID','Customer_Name':'Customer Name'}, inplace=True)

# Sort in descending order
top_spenders = customer_sales.sort_values(by='Sales', ascending=False)

print(top_spenders.head(10).reset_index(drop=True))
```

## Sale Trade Analysis

```python

yearly_sales = df.groupby(df['Order_Date'].dt.year)['Sales'].sum()

# setting new index and reaming the columns
yearly_sales = yearly_sales.reset_index()
yearly_sales = yearly_sales.rename(columns={'Order_Date':'Year', 'Sales': 'Total Sales'})

print(yearly_sales)
```

```python
#Plotting a bar graph
plt.bar(yearly_sales['Year'], yearly_sales['Total Sales'])

# Set labels
plt.title('Yearly Sales')
plt.xlabel('Year')
plt.ylabel('Total Sales')

plt.xticks(rotation=65)

plt.show()
```

```python
#Plotting a line graph

plt.plot(yearly_sales['Year'], yearly_sales['Total Sales'], marker='o', linestyle='--')

# Set labels
plt.title('Yearly Sales')
plt.xlabel('Years')
plt.ylabel('Total Sales')

plt.xticks(rotation=65)

plt.show()
```

# Quartely Sales

```python
# filter data according to year
year_sales = df[df['Order_Date'].dt.year == 2017]

# Calculate quartely sales for year 2017
quartely_sales = year_sales.resample('QE', on='Order_Date')['Sales'].sum()

quartely_sales =quartely_sales.reset_index()
quartely_sales = quartely_sales.rename(columns={'Order_Date': 'Quarter', 'Sales':'Total Sales'})

print('This are the Quarterly sales for 2017')
print(quartely_sales)
```

```python
#Plotting a line graph

plt.plot(quartely_sales['Quarter'], quartely_sales['Total Sales'], marker='o', linestyle='--')

# Set labels
plt.title('Quarter Sales')
plt.xlabel('Quarters')
plt.ylabel('Total Sales')

plt.xticks(rotation=65)

plt.show()
```

## Monthly sales trend for a year

```python
year_sales = df[df['Order_Date'].dt.year == 2018]

# Filter data according to year
monthly_sales = year_sales.resample('ME', on='Order_Date')['Sales'].sum()

# calculate monthly sales for year 2017
monthly_sales = monthly_sales.reset_index()
monthly_sales = monthly_sales.rename(columns={'Order_Date': 'Month', 'Sales':'Total Monthly Sales'})

print('This are the Monthky sales for 2018')
print(monthly_sales)
```

```python
#Plotting a line graph

plt.plot(monthly_sales['Month'], monthly_sales['Total Monthly Sales'], marker='o', linestyle='-')

# Set labels
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')

plt.xticks(rotation=65)

plt.show()
```

## Sales Trends

```python
# Group the data by months and calculate the total sales amount for each month
monthly_sales = df.groupby(df['Order_Date'].dt.to_period('M'))['Sales'].sum()

# Plot the sales trends for months
plt.figure(figsize=(12, 26))

# Monthly Sales Trend
plt.subplot(3, 1, 1)
monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales Amount')

# Adjust layout and display the plots
# plt.tight_layout()
plt.show()
```

```python
# Convert the "Order Date" column to datetime format
df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True)

# Group the data by quarters and calculate the total sales amount for each quarter
quarterly_sales = df.groupby(df['Order_Date'].dt.to_period('Q'))['Sales'].sum()

# Plot the sales trends for months, quarters, and years
plt.figure(figsize=(12, 20))

# Quarterly Sales Trend
plt.subplot(3, 1, 2)
quarterly_sales.plot(kind='line', marker='o')
plt.title('Quarterly Sales Trend')
plt.xlabel('Quarter')
plt.ylabel('Sales Amount')

# Adjust layout and display the plots
#plt.tight_layout()
plt.show()
```
